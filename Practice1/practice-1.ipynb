```python
class Question:
    def __init__(self, prompt, answer):
        self.prompt = prompt
        self.answer = answer

question_prompts = [
	"What color are Apples?\n(a) Red/Green\n(b) Purple\n(c) Orange\n\n",
	"What color are Bananas?\n(a) Teal\n(b) Magenta\n(c) Yellow\n\n",
	"What color are Straberries?\n(a) Yellow\n(b) Red\n(c) Blue\n\n"
]

questions = [
	Question(question_prompts[0], "a"),
	Question(question_prompts[1], "c"),
	Question(question_prompts[2], "b"),
]

def run_test(questions):
	score = 0
	for question in questions:
		answer = input(question.prompt)
		if answer == question.answer:
			score += 1
	print("You got " + str(score) + "/" + str(len(questions)) + " correct")
	
run_test(questions)
```

```python
def fib(n):    # write Fibonacci series less than n
    """Print a Fibonacci series less than n."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

# Now call the function we just defined:
#fib(100)
f = fib # new call func
f(100)
```

```python
from enum import Enum

class Color(Enum):
    RED = 'red'
    GREEN = 'green'
    BLUE = 'blue'

# Get input and ensure it matches a valid enum
user_input = input("Enter your choice of 'red', 'blue' or 'green': ").lower()

# Convert the input string to a Color enum member
try:
    color = Color(user_input)
    match color:
        case Color.RED:
            print("I see red!")
        case Color.GREEN:
            print("Grass is green")
        case Color.BLUE:
            print("I'm feeling the blues :(")
except ValueError:
    print("Invalid color choice! Please enter 'red', 'blue', or 'green'.")
```

```python
def ask_ok(prompt, retries=4, reminder='Please try again!'):
    while True:
        reply = input(prompt)
        if reply in {'y', 'ye', 'yes'}:
            return True
        if reply in {'n', 'no', 'nop', 'nope'}:
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)

ask_ok('Do you really want to quit?')
```

```python
import matplotlib.pyplot as plt

# Define audience categories
categories = [
    "Busy Professionals & Working Parents",
    "Health-Conscious Consumers",
    "Budget-Conscious Shoppers",
    "Students & Young Adults",
    "Elderly & Forgetful Shoppers"
]

# Define proportions
proportions = [30, 10, 20, 10, 30]  

# Define colors
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"]

# Create the figure
fig, ax = plt.subplots(figsize=(8, 6))

# Create pie chart
ax.pie(proportions, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Set title
ax.set_title("Primary Target Audience Distribution")

# Save the figure
fig.savefig('TargetAudience_PieChart.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define audience categories
categories = [
    "Tech-Savvy Individuals & Mobile App Users",
    "Meal Planners & Home Cooks",
    "Budget-Conscious Households & Waste Reducers",
    "Frequent Grocery Shoppers"
]

proportions = [25, 25, 25, 25]  

# Define colors
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create bar chart
ax.bar(categories, proportions, color=colors, edgecolor='black')
ax.set_xlabel("Audience Categories")
ax.set_ylabel("Proportion")
ax.set_title("Customer Segments")
ax.set_xticks(range(len(categories)))
ax.set_xticklabels(categories, rotation=45, ha='right')

# Adjust layout to prevent cropping
plt.tight_layout()

# Save the correct figure
fig.savefig('CustomerSegments.png', bbox_inches='tight')

# Show the plot
plt.show()
```
